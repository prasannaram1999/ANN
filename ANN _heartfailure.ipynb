```python
def remove(df,column,threshold):
    sns.boxplot(df[column])
    plt.show()
    removed_outliers  = df[df[column]<= threshold]
    sns.boxplot(removed_outliers[column])
    plt.show()
    return removed_outliers
threshold = 400000
remove(df,'platelets',threshold)
    
```

```python
X = df.drop(['DEATH_EVENT'],axis=1)
y = df['DEATH_EVENT']
```

```python
df['DEATH_EVENT'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
X = sd.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from keras import callbacks
early_stopping = callbacks.EarlyStopping(min_delta=0.001,patience=20,restore_best_weights=True)
```

```python
model = Sequential()
```

```python
model.add(Dense(50,kernel_initializer='uniform',input_dim=12,activation='relu'))
```

```python
model.add(Dense(20,kernel_initializer='uniform',activation='relu'))
model.add(Dense(16,kernel_initializer='uniform',activation='relu'))
model.add(Dense(10,kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(2,kernel_initializer='uniform', activation='sigmoid'))
```

```python
from keras.optimizers import SGD
model.compile(loss='crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=32,epochs=15,callbacks=[early_stopping],validation_split=0.2)
```

```python
model.build()
```

```python
model.summary()
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
np.set_printoptions()
```
